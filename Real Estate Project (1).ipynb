```python
#QUESTION 10:
#10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less,
#it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, 
#it should be medium, otherwise it should be considered large

import numpy as np
conditions = [
              (df['total_bedrooms'] <=10),
              (df['total_bedrooms'] >=11)&(df['total_bedrooms'] <=1000),
              (df['total_bedrooms'] > 1000)
             ]
values = ['small','medium','large']
df['total_bedroom_size']=np.select(conditions,values)
df

#Here mainly condition with respect to total_bedrooms are applied and 
#the data is categorized as small,medium,big and a seperate column is created
#at the end of table.Hence there is increase in number of columns.
```
